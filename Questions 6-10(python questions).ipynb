```python
# Question 6
def pig(word):
    vowels = ("a","e","i","o","u")
    if word.startswith(vowels):
        return word +"way"
    else:
        first_char = word[0]
        newword = word[1:]
        final_word = newword + first_char

        return final_word + "ay"


user_input = input("Enter your word / sentence to convert: ")

print("the PigLatin form is: " + pig(user_input))
```

```python
#Question 8
def curconv(currency, amount):
    curdict = {}
    with open("currencies.txt") as myfile:
        for line in myfile:
            words =line.split( ) # split lines into tuples
            curdict[words[0]]= words[1]# add tuple values to dictionary

            if currency == words[0]:
                 rate = float(curdict[currency])
                #  print("rate =", curdict[currency])
                 newamount = int(amount) * rate
                 return newamount
   
    
cur = input("Enter the currency you wish to convert: ")
amount = input ("Enter the amount: ")
currencies = ["AUD","CHF","CNY","DKK","EUR","GBP","HKD","INR","JPY","MXN","MYR","NOK","NZD","PHP","SEK","SGD","THB"]

if cur in currencies:
    print (cur + " "+amount + " is USD "+str(curconv(cur,amount)))
else:
    print("Invalid currency, Please input a correct currency notation in CAPITAL letters")
```

```python
#Question 10
def frequencies(word):
    letters = [0] * 26 #create array to store the encrpyted message
    for character in word:
        if character.isalpha(): # check only for alphabets
            character_in_array = ord(character.lower())-97 # get ascii values
            letters[character_in_array] = letters[character_in_array]+1 #count each character in the message
    return letters

user_input = input("Encrypt message: ")
print("Result: ",frequencies(user_input))
```
