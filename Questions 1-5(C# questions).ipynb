```
// Question 4
Console.WriteLine("Input the number of elements to be stored: ");
int arrlen = Convert.ToInt32(Console.ReadLine());
int[] nums = new int[arrlen];
int m = 0, elen = 0, olen = 0, j = 0, k = 0;
Console.WriteLine("Enter the elements of the array: ");
for (int i = 0; i < arrlen; i++)
{
    nums[i] = Convert.ToInt32(Console.ReadLine());
}
while (m < arrlen)
{

    if (nums[m] % 2 == 0)
    {
        elen++;
    }
    else if (nums[m] % 2 == 1)
    {
        olen++;
    }
    m++;
}
int[] even = new int[elen];
int[] odds = new int[olen];

// seperate the odd and even elements from the initial array into seperate arrays
for (int i = 0; i < arrlen; i++)
{
    if (nums[i] % 2 == 0)
    {
        even[j] = nums[i];
        j++;
    }
    else
    {
        odds[k] = nums[i];
        k++;
    }
}

//print even numbers
Console.WriteLine("The even numbers are: ");
foreach (int i in even)
{
    Console.WriteLine(i);
}

//print odd numbers
Console.WriteLine("The odd numbers are");
foreach (int i in odds)
{
    Console.WriteLine(i);
}
```

```
//question 5a
static bool inside(double x1, double y1, double x2,
                      double y2, double x, double y)
{
    if (x > x1 && x < x2 &&
        y > y1 && y < y2)
        return true;

    return false;
}

Console.WriteLine("=========================================Question 5a===================================================\n");
// bottom-left and top-right
// corners of rectangle

Console.WriteLine("Enter the lower left corner points: ");
double x1 = Convert.ToDouble(Console.ReadLine());
double y1 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the lupper right corner points: ");
double x2 = Convert.ToDouble(Console.ReadLine());
double y2 = Convert.ToDouble(Console.ReadLine());

// given point
Console.WriteLine("Enter the point to check: ");
double x = Convert.ToDouble(Console.ReadLine());
double y = Convert.ToDouble(Console.ReadLine());

Console.WriteLine($"The point ({x},{y}), lie in the rectangle (({x1},{y1}), ({x2},{y2}))?");

// function call
Console.WriteLine(inside(x1, y1, x2, y2, x, y));


//question 5b
Console.WriteLine("\n=======================================Question 5b===================================================\n");
Console.WriteLine("Enter the lower left corner points of first rectangle: ");
double a1 = Convert.ToDouble(Console.ReadLine());
double b1 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the lupper right corner points of first rectangle: ");
double a2 = Convert.ToDouble(Console.ReadLine());
double b2 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the lower left corner points of second rectangle: ");
double a3 = Convert.ToDouble(Console.ReadLine());
double b3 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the lupper right corner points of second rectangle: ");
double a4 = Convert.ToDouble(Console.ReadLine());
double b4 = Convert.ToDouble(Console.ReadLine());

// given point
Console.WriteLine("Enter the point to check: ");
double a = Convert.ToDouble(Console.ReadLine());
double b = Convert.ToDouble(Console.ReadLine());


// function call
bool ans1 = inside(a1, b1, a2, b2, a, b);

bool ans2 = inside(a3, b3, a4, b4, a, b);

if (ans1 == ans2)
{
    Console.WriteLine($"The point ({a},{b}), lies in the both rectangles (({a1},{b1}), ({a2},{b2})) and (({a3},{b3}),({a4},{b4}))");
}
else
{
    Console.WriteLine($"The point ({a},{b}), does not lie in the both rectangles (({a1},{b1}), ({a2},{b2})) and (({a3},{b3}),({a4},{b4}))");
}
```
